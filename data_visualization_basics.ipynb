### Colour Each Dot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

### ColorMap

```python
colors = np.array([0,10,20,30,40,50,55,60,65, 70,85,90,100])
```

```python
plt.scatter(x,y, c = colors, cmap = 'viridis')
plt.colorbar()
plt.show()
```

# Bar Plot

```python
X = np.arange(10)
Y = np.random.uniform(0,1,10)

fig, ax = plt.subplots()
ax.bar(X,Y)
```

## Horizontal Bar Plot

```python
X = np.arange(10)
Y = np.random.uniform(0,1,10)

fig, ax = plt.subplots()
ax.barh(X,Y)
```

##### Bar Color

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

##### Bar Width

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red", width=0.5)
plt.show()
```

#### Horizontal Plot Height

```python
X = np.arange(10)
Y = np.random.uniform(0,1,10)

fig, ax = plt.subplots()
ax.barh(X,Y, height=0.5)
```

# PieChart

```python
Z = np.random.uniform(0,1,8)
plt.pie(y)
plt.show()
```

### Labels

```python
Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
plt.pie(Z, labels = label)
plt.show()
```

### Start Angle

```python
Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
plt.pie(Z, labels = label, startangle=180)
plt.show()
```

### Explode

```python
Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
explode=[0.1,0,0,0,0]
plt.pie(Z, labels = label, startangle=180, explode=explode)
plt.show()
```

### Shadow

```python
Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
explode=[0.1,0,0.5,0,0]
plt.pie(Z, labels = label, startangle=180, explode=explode, shadow=True)
plt.show()
```

### Colors

```python
Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
color = ['red','orange','yellow','green','pink']
explode=[0.1,0,0.5,0,0]
plt.pie(Z, labels = label, colors = color,startangle=180, explode=explode, shadow=True)
plt.show()
```

### Legend

```python
Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
color = ['red','orange','yellow','green','pink']
explode=[0.1,0,0.5,0,0]
plt.pie(Z, labels = label, colors = color,startangle=180, explode=explode, shadow=True)
plt.legend()
plt.show()
```
